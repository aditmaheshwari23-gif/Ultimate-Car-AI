```python
import pandas as pd
import re

df = pd.read_csv("cars.csv")

# clean
df["brand"] = df["brand"].astype(str).str.lower().str.strip()
df["fuel"] = df["fuel"].astype(str).str.lower().str.strip()

print("🚗 ULTIMATE CAR AI READY")
print("Try:")
print("👉 bmw under 5000000")
print("👉 toyota vs honda")
print("👉 cheap cars")
print("👉 diesel cars")
print("👉 exit")


# extract number (price)
def extract_price(text):
    nums = re.findall(r'\d+', text)
    return int(nums[0]) if nums else None


# extract brands
def get_brands(text):
    found = []
    for b in df["brand"].unique():
        if b in text:
            found.append(b)
    return found


while True:
    q = input("\nAsk: ").lower().strip()

    if q == "exit":
        print("👋 bye")
        break

    result = df.copy()

    # ---------------- COMPARE MODE ----------------
    if "vs" in q or "compare" in q:
        brands = get_brands(q)

        if len(brands) < 2:
            print("❌ Please mention TWO brands (e.g. toyota vs honda)")
            continue

        b1, b2 = brands[0], brands[1]

        d1 = df[df["brand"] == b1]
        d2 = df[df["brand"] == b2]

        print("\n🆚 COMPARISON\n")

        print(f"🚗 {b1.upper()} | Avg Price:", round(d1["selling_price"].mean(), 2))
        print(f"🚗 {b2.upper()} | Avg Price:", round(d2["selling_price"].mean(), 2))

        print("\n🏆 Winner (cheaper):",
              b1 if d1["selling_price"].mean() < d2["selling_price"].mean() else b2)

        continue

    # ---------------- PRICE FILTER ----------------
    price = extract_price(q)
    if "under" in q and price:
        result = result[result["selling_price"] <= price]

    # ---------------- BRAND FILTER ----------------
    brands = get_brands(q)
    if brands:
        result = result[result["brand"].isin(brands)]

    # ---------------- FUEL FILTER ----------------
    if "diesel" in q:
        result = result[result["fuel"] == "diesel"]

    if "petrol" in q:
        result = result[result["fuel"] == "petrol"]

    # ---------------- SMART SORTING ----------------
    if "cheap" in q:
        result = result.sort_values("selling_price").head(10)

    if "expensive" in q:
        result = result.sort_values("selling_price", ascending=False).head(10)

    # ---------------- OUTPUT ----------------
    if len(result) == 0:
        print("❌ No results found")
    else:
        print("\n📊 RESULTS:")
        print(result.head(10))

        print("\n🧠 INSIGHT:")
        print("Average price:", round(result["selling_price"].mean(), 2))
        print("Total cars:", len(result))
```
